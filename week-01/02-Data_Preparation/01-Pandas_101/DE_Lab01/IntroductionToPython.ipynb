## Data Structures: Lists

```python
my_list = [1, 2, 3]
my_list
```

```python
my_list += [4, 5] # my_list = my_list + [4, 5]
my_list
```

```python
my_list.append(6)
my_list
```

```python
print(my_list[0])
print(my_list[1])
print(my_list[-1])
print(my_list[1:3])
print(my_list[1:])
print(my_list[:3])
```

```python
# for loop
for item in my_list:
    print(item)
```

```python
# while loop
sum_list = 0
count = 0
while sum_list < 20:
    sum_list += my_list[count]
    count += 1
    print(f"Sum after {count} iterations: {sum_list}")
```

## Data Structures: Dicts

```python
my_dict = {'key 1': 'value 1', 'key 2': 'value 2', 'key 3': 'value 3', 'key 4': 'value 4'}
```

```python
my_dict['key 2'] = 12
del my_dict['key 4']
my_dict['key 5'] = 'value 5'
```

```python
for k,v in my_dict.items():
    print(f"Value for key {k}: {v}")
```

```python
# for loop with enummerate wrapper
for i, (k, v) in enumerate(my_dict.items()):
    print(f"{i} - Value for key {k}: {v}")
```

## More Control Flow Tools

```python
for i in range(10):
    print(i)
```

```python
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
for i in range(10):
    if i>5:
        break
    print(i)
```

```python
a = 3

# if-else statement
if a > 2:
    b = 1
else:
    b = 2
print(b)
```

```python
# same in one line:
b = 1 if a > 2 else 2
print(b)
```

```python
# List Comprehension
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = [x for x in fruits if "a" in x]

print(newlist)
```

```python
# also works for dicts
my_dict = {1:2, 2:4, 3:9}
my_dict2 = {key:value+1 for (key,value) in my_dict.items()}
print(my_dict2)
```

## Defining Functions
